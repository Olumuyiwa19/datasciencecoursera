```
 # replace "?" to NaN
df_CarAttributes.replace("?", np.nan, inplace = True)
df_CarAttributes.head(10)
```

```
missing_data = df_CarAttributes.isnull()
missing_data.head()
```

```
#To count missing value in each column
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

```
#To resolve the missing values we find average of each column and replace their missing values with the average

#we find the average normal loss in order to replace missing data for normalized-loses
avg_normal_loss = df_CarAttributes["NORMALIZED-LOSSES"].astype("float").mean(axis=0)
print('Average of NORMALIZED-LOSSES:', avg_normal_loss)

#replace the missing value with the average normal loss
df_CarAttributes["NORMALIZED-LOSSES"].replace(np.nan,avg_normal_loss,inplace=True)

#we find the average normal loss in order to replace missing data for bore
avg_bore = df_CarAttributes['BORE'].astype('float').mean(axis=0)
print('Average of Bore:',avg_bore)

#replace the missing value with the average bore
df_CarAttributes['BORE'].replace(np.nan,avg_bore,inplace=True)

#we find the average of stroke  in order to subtitute for missing values in its columns
avg_stroke = df_CarAttributes['STROKE'].astype('float').mean(axis=0)
print('Average of Stroke:',avg_stroke)

#replace the missing value with the average stroke
df_CarAttributes['STROKE'].replace(np.nan,avg_stroke,inplace=True)

#convert the following dataset from object types to the float types; compression-ratio, peak-rpm, 
df_CarAttributes[["PEAK-RPM","COMPRESSION-RATIO"]] = df_CarAttributes[["PEAK-RPM","COMPRESSION-RATIO"]].astype("float")

#we find the average of compresion-ratio  in order to subtitute for missing values in its columns
avg_compression_ratio = df_CarAttributes['COMPRESSION-RATIO'].mean(axis=0)
print('Average of Compression-Ratio:', avg_compression_ratio)

#replace the missing value with the average compression ratio
df_CarAttributes['COMPRESSION-RATIO'].replace(np.nan,avg_compression_ratio,inplace=True)

#we find the average of peak-rpm  in order to subtitute for missing values in its columns
avg_rpm = df_CarAttributes['PEAK-RPM'].mean(axis=0)
print('Average of PEAK-RPM:', avg_rpm)

#replace the missing value with the average peak rpm
df_CarAttributes['PEAK-RPM'].replace(np.nan,avg_rpm,inplace=True)


# check your transformed data
df_CarAttributes.head()
```

```
# standardizing some of the attributes. we convert mpg to L/100km for city and highway by mathematical operation (235 divided by mpg)

df_CarAttributes['CITY-L/100KM'] = 235/df_CarAttributes['CITY-MPG']
df_CarAttributes['HIGHWAY-L/100KM'] = 235/df_CarAttributes['HIGHWAY-MPG']

# check your transformed data
df_CarAttributes.head(2)
```

```
#To normalized the length, width and height measurement we replace (original value) by (original value)/(maximum value) so that their value ranges between 0 and 1

df_CarAttributes['LENGTH'] = df_CarAttributes['LENGTH']/df_CarAttributes['LENGTH'].max()
df_CarAttributes['WIDTH'] = df_CarAttributes['WIDTH']/df_CarAttributes['WIDTH'].max()
df_CarAttributes['HEIGHT'] = df_CarAttributes['HEIGHT']/df_CarAttributes['HEIGHT'].max()
df_CarAttributes.head()
```

```
#To carry out binning operation we first drop all the NaN in the horsepower column
df_CarAttributes = df_CarAttributes.dropna(subset =['HORSEPOWER'])

#Change the format from object to integer
df_CarAttributes["HORSEPOWER"]=df_CarAttributes["HORSEPOWER"].astype(int, copy=True)

#We plot the histogram of horspower, to see what the distribution of horsepower looks like.
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df_CarAttributes["HORSEPOWER"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```

""" 
We carry out Binning operation which is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis
We would like 3 bins of equal size bandwidth so we use numpy's linspace(start_value, end_value, numbers_generated function.
Since we want to include the minimum value of horsepower we want to set start_value=min(df["horsepower"]). 
Since we want to include the maximum value of horsepower we want to set end_value=max(df["horsepower"]).
Since we are building 3 bins of equal length, there should be 4 dividers, so numbers_generated=4.

"""

#We build a bin array, with a minimum value to a maximum value, with bandwidth calculated above. The bins will be values used to determine when one bin ends and another begins.
bins = np.linspace(min(df_CarAttributes["HORSEPOWER"]), max(df_CarAttributes["HORSEPOWER"]), 4)
bins



```

```
#we set the group name
group_names = ['Low', 'Medium', 'High']

#We apply the function "cut" to determine what each value of "df_CarAttributes["HORSEPOWER"]" belongs to.
df_CarAttributes['HORSEPOWER-BINNED'] = pd.cut(df_CarAttributes["HORSEPOWER"], bins, labels=group_names, include_lowest=True )

#we view the transformed result
df_CarAttributes[['HORSEPOWER','HORSEPOWER-BINNED']].head(20)
```

```
#we count the number of bins the vehincle fall into
df_CarAttributes["HORSEPOWER-BINNED"].value_counts()
```

```
#we consider to plot the distribution of each bin.
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df_CarAttributes["HORSEPOWER-BINNED"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
